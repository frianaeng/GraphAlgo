```python
from MovieNetwork import MovieNetwork
import pickle
import pandas as pd
import networkx

with open("data/pkl/criterion_movie_network_full.pkl", "rb") as f:
    movie_net = pickle.load(f)
```

```python
#type(movie_net)
#dir(movie_net)
movie_net.adjacency
```

```python
nodes = [
    {"id": n, "label": n}
    for n in movie_net.nodes
]

links = [
    {"source": u, "target": v}
    for u, v in movie_net.edges
]

#print(nodes)
print(links)
```

```python
graph_json = {
    "nodes": nodes,
    "links": links
}
import json

with open("movienetwork_unweighted.json", "w") as f:
    json.dump(graph_json, f, indent=2)

```
