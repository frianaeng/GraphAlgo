## Data wrangling 

```python
from collections import defaultdict
import pickle
import os

class MovieNetwork:
    def __init__(self):
        """Creates a movie similarity network"""
        # adjacency: movie -> {other_movie : weight}
        self.adjacency = defaultdict(dict)
        self.features = {}

    def getNumMovies(self):
        return len(self.adjacency)

    def getNumEdges(self):
        numEdges = 0
        for _, nbrs in self.adjacency.items():
            numEdges += len(nbrs)
        return int(numEdges / 2)

    def addMovie(self, movie, feature_set):
        """Add a movie and its feature set"""
        if movie not in self.adjacency:
            self.adjacency[movie] = {}
        self.features[movie] = set(feature_set)

    def addEdge(self, movie1, movie2, weight):
        """Add an undirected weighted edge"""
        self.adjacency[movie1][movie2] = weight
        self.adjacency[movie2][movie1] = weight


    def jaccard(self, s1, s2):
        if len(s1 | s2) == 0:
            return 0.0
        return len(s1 & s2) / len(s1 | s2)

    def buildEdges(self, threshold=0.0):
        """Compute Jaccard weights for all movie pairs"""
        movies = list(self.features.keys())

        for i in range(len(movies)):
            for j in range(i + 1, len(movies)):
                m1, m2 = movies[i], movies[j]
                w = self.jaccard(self.features[m1], self.features[m2])
                if w > threshold:
                    self.addEdge(m1, m2, w)


    def DFS(self, start_movie):
        start = time.time()
    
        stack = []
        stack.append(start_movie)
    
        parent = {}
        visited = {}
    
        # initialize
        for movie in self.adjacency:
            visited[movie] = False
            parent[movie] = None
    
        visited[start_movie] = True
    
        while len(stack) != 0:
            u = stack.pop()
    
            # u's neighbors are movies
            for v in self.adjacency[u]:
                if not visited[v]:
                    visited[v] = True
                    parent[v] = u
                    stack.append(v)
    
        end = time.time()
        print(end - start)
    
        return parent


    def BFS(self, start):
        import queue
        Q = queue.SimpleQueue()
        Q.put(start)

        visited = {m: False for m in self.adjacency}
        parent  = {m: None  for m in self.adjacency}

        visited[start] = True

        while not Q.empty():
            u = Q.get()
            for v in self.adjacency[u]:
                if not visited[v]:
                    visited[v] = True
                    parent[v] = u
                    Q.put(v)
        return parent

    def countConnectedComponents(self):
        visited = {m: False for m in self.adjacency}
        components = 0

        for movie in self.adjacency:
            if not visited[movie]:
                components += 1
                stack = [movie]
                visited[movie] = True

                while stack:
                    u = stack.pop()
                    for v in self.adjacency[u]:
                        if not visited[v]:
                            visited[v] = True
                            stack.append(v)
        return components

    def getConnectedComponents(self):
        visited = {m: False for m in self.adjacency}
        components = []
    
        for movie in self.adjacency:
            if not visited[movie]:
                stack = [movie]
                visited[movie] = True
                component = []
    
                while stack:
                    u = stack.pop()
                    component.append(u)
                    for v in self.adjacency[u]:
                        if not visited[v]:
                            visited[v] = True
                            stack.append(v)
    
                components.append(component)
    
        return components



    
    # -----------------------------
    # Similarity-specific queries
    # -----------------------------

    def mostSimilar(self, movie, k=5):
        """Return top-k most similar movies"""
        return sorted(
            self.adjacency[movie].items(),
            key=lambda x: x[1],
            reverse=True
        )[:k]

    # -----------------------------
    # Persistence
    # -----------------------------

    def save(self, path):
        os.makedirs(os.path.dirname(path), exist_ok=True)
        with open(path, "wb") as f:
            pickle.dump(self, f)


            

#A very high Jaccard weight means the two movies share most of their features and are very similar under 
#the chosen feature set.


    def loadDataset(self, filename):
        """
        Loads movie data and builds a movie similarity network
        based on Jaccard similarity of feature sets.
        """
        import pickle
    
        with open(filename, 'rb') as f:
            data = pickle.load(f)
    
        # data expected format:
        # movie -> iterable of features (e.g., actors, genres, etc.)
    
        # Step 1: add movies and feature sets
        for movie, features in data.items():
            self.addMovie(movie, features)
    
        movies = list(self.features.keys())
    
        # Step 2: compute pairwise Jaccard similarity
        for i in range(len(movies)):
            for j in range(i + 1, len(movies)):
                m1, m2 = movies[i], movies[j]
                s1, s2 = self.features[m1], self.features[m2]
    
                sim = self.jaccard(s1, s2)
                if sim > 0:
                    self.addEdge(m1, m2, sim)


    def visualize_network(self):
        """Visualize the movie similarity network using NetworkX and Matplotlib"""
        import networkx as nx
        import matplotlib.pyplot as plt
    
        G = nx.Graph()
    
        # Add nodes (movies)
        for movie in self.adjacency:
            G.add_node(movie)
    
        # Add edges (movie similarity, weighted)
        for movie, neighbors in self.adjacency.items():
            for other_movie, weight in neighbors.items():
                if not G.has_edge(movie, other_movie):
                    G.add_edge(movie, other_movie, weight=weight)
    
        # Layout
        pos = nx.circular_layout(G)
    
        # Draw graph
        plt.figure(figsize=(10, 8))
        nx.draw(
            G,
            pos,
            with_labels=True,
            node_color="lightgreen",
            node_size=500,
            font_size=8,
            font_weight="bold",
            edge_color="gray"
        )
    
        plt.title("Movie Similarity Network Visualization")
        plt.show()
    
```

```python
import pickle
import types
import sys
from collections import defaultdict

def load_criterion_movie_features(path, num_movies=1805):
    """
    Returns:
        movie_features: dict[movie_id] -> set(actors)
    """

    # Stub ActorNetwork so pickle loads cleanly
    module = types.ModuleType("ActorNetwork")
    class ActorNetwork:
        pass
    module.ActorNetwork = ActorNetwork
    sys.modules["ActorNetwork"] = module

    with open(path, "rb") as f:
        actor_net = pickle.load(f)

    movie_features = defaultdict(set)

    # actor_net.movies: actor -> set(movie_ids)
    for actor, movie_ids in actor_net.movies.items():
        for m in movie_ids:
            movie_features[m].add(actor)

    # Ensure full 0..1804 movie index space exists
    for m in range(num_movies):
        movie_features[m]  # touch key

    return movie_features
```

```python
#from movie_network import MovieNetwork  # or wherever your class lives

CRITERION_PATH = "data/pkl/criterion_actor_network.pkl"

# Step 1: load movie -> actor features
movie_features = load_criterion_movie_features(CRITERION_PATH)

# Step 2: create MovieNetwork
MN = MovieNetwork()

# Step 3: add movies + feature sets
for movie_id, actors in movie_features.items():
    MN.addMovie(movie_id, actors)

print("Movies added:", MN.getNumMovies())

# Step 4: build edges using Jaccard similarity
# threshold=0 keeps all shared-actor connections
MN.buildEdges(threshold=0.0)

print("Edges added:", MN.getNumEdges())
print("Connected components:", MN.countConnectedComponents())
```

```python
components = MN.getConnectedComponents()

print("Total components:", len(components))

sizes = sorted(len(c) for c in components)
print("Smallest sizes:", sizes[:20])
print("Largest size:", sizes[-1])
```

```python
small_components = [c for c in components if 2 <= len(c) <= 5]

comp = small_components[0]
print("Component:", comp)
```

```python
actors_in_component = set()
for m in comp:
    actors_in_component |= movie_features[m]

print("Actors in component:", actors_in_component)
```

```python
leaks = {}

for m, actors in movie_features.items():
    if m not in comp:
        overlap = actors & actors_in_component
        if overlap:
            leaks[m] = overlap

if not leaks:
    print("Actors are isolated to this component")
else:
    print("Actor leakage detected:")
    for m, a in list(leaks.items())[:10]:
        print("Movie", m, "shares actors", a)



```

```python
degrees = [len(MN.adjacency[m]) for m in MN.adjacency]

print("Degree 0 movies:", sum(d == 0 for d in degrees))
print("Degree 1 movies:", sum(d == 1 for d in degrees))
print("Average degree:", sum(degrees)/len(degrees))
```

```python
components = MN.getConnectedComponents()

singletons = [c[0] for c in components if len(c) == 1]

print("Number of singleton nodes:", len(singletons))
singleton_actors = {}

for m in singletons:
    singleton_actors[m] = movie_features[m]
leaks = {}

for singleton_movie, actors in singleton_actors.items():
    for m, other_actors in movie_features.items():
        if m != singleton_movie:
            overlap = actors & other_actors
            if overlap:
                leaks.setdefault(singleton_movie, set()).update(overlap)
```

```python
if not leaks:
    print("All singleton movies are truly isolated.")
    print("No actor in a singleton movie appears in any other movie.")
else:
    print("Actor leakage detected in singleton movies:")
    print("Number of problematic singleton movies:", len(leaks))
    for m, a in list(leaks.items())[:10]:
        print("Movie", m, "shares actors", a)

```

```python
#6a
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx
# =========================
# Build adjacency matrix from network.adjacency
# =========================

movies = list(network.adjacency.keys())
idx = {movie: i for i, movie in enumerate(movies)}

n = len(movies)
A = np.zeros((n, n))

for m1, neighbors in network.adjacency.items():
    i = idx[m1]
    for m2, weight in neighbors.items():
        j = idx[m2]
        A[i, j] = weight

print("Number of nodes:", n)
print("Number of edges:", int(np.count_nonzero(A) / 2))

# =========================
# Build graph from adjacency matrix
# =========================

G = nx.from_numpy_array(A)
# =========================
# Graph visualization
# =========================

plt.figure(figsize=(10, 10))

pos = nx.spring_layout(
    G,
    seed=42,
    k=1 / np.sqrt(n)   # improves spacing for larger graphs
)

nx.draw_networkx_nodes(
    G,
    pos,
    node_size=20,
    node_color="black",
    alpha=0.8
)

nx.draw_networkx_edges(
    G,
    pos,
    width=0.3,
    alpha=0.3
)

plt.title("Movie Network Visualization (Spring Layout)")
plt.axis("off")
plt.show()
```

```python
#6b
# =========================
# Degree matrix
# =========================

degrees = A.sum(axis=1)
D_inv_sqrt = np.diag(1.0 / np.sqrt(degrees))
# =========================
# Normalized Laplacian
# =========================

L_sym = np.eye(n) - D_inv_sqrt @ A @ D_inv_sqrt
# =========================
# Heatmap visualization
# =========================

plt.figure(figsize=(8, 8))

# mask diagonal so off-diagonal structure is visible
L_masked = L_sym.copy()
np.fill_diagonal(L_masked, np.nan)

plt.imshow(L_masked, cmap="Blues", interpolation="nearest")
plt.colorbar(label="Laplacian Value")

plt.title("Normalized Graph Laplacian Heatmap")
plt.xlabel("Node Index")
plt.ylabel("Node Index")

plt.show()
```

```python
#6c

# =========================
# Safety: rebuild L_sym robustly (handles isolated nodes)
# =========================
degrees = A.sum(axis=1)

inv_sqrt = np.zeros_like(degrees, dtype=float)
mask = degrees > 0
inv_sqrt[mask] = 1.0 / np.sqrt(degrees[mask])

D_inv_sqrt = np.diag(inv_sqrt)
L_sym = np.eye(n) - D_inv_sqrt @ A @ D_inv_sqrt

# Clean numerical noise (optional but helps)
L_sym = 0.5 * (L_sym + L_sym.T)

# =========================
# Eigenvalues / eigenvectors of normalized Laplacian
# =========================
eigvals, eigvecs = np.linalg.eigh(L_sym)  # symmetric => stable, sorted ascending

print("Smallest 10 eigenvalues:")
print(eigvals[:10])

# =========================
# Plot first 20 eigenvalues
# =========================
k_plot = min(20, n)
plt.figure(figsize=(8, 4))
plt.plot(range(1, k_plot + 1), eigvals[:k_plot], marker="o")
plt.title("First 20 Eigenvalues of Normalized Laplacian")
plt.xlabel("Index (sorted)")
plt.ylabel("Eigenvalue")
plt.grid(True, alpha=0.3)
plt.show()

# =========================
# Spectral gaps: look at gaps among the first ~50 eigenvalues
# =========================
k_check = min(50, n)
gaps = np.diff(eigvals[:k_check])   # gap[i] = eigvals[i+1] - eigvals[i]

# Show top few gaps (largest jumps)
top = min(10, len(gaps))
gap_order = np.argsort(gaps)[::-1][:top]

print("\nLargest spectral gaps among first", k_check, "eigenvalues:")
for i in gap_order:
    print(f"gap between [{i}]={eigvals[i]:.6f} and [{i+1}]={eigvals[i+1]:.6f}  ->  {gaps[i]:.6f}")

# Interpretation helper:
# If there is a large gap after [k-1], that often suggests ~k communities.
# (Look for a big jump among the small eigenvalues near 0.)
tol = 1e-8
num_zero = int(np.sum(eigvals < tol))
print("\nApprox # of zero eigenvalues (connected components):", num_zero)
```

```python
#6d

# =========================
# Extract Fiedler vector robustly
# =========================
tol = 1e-8
fiedler_index = int(np.where(eigvals > tol)[0][0])  # first nonzero eigenvalue
fiedler = eigvecs[:, fiedler_index]

print("Fiedler eigenvalue:", eigvals[fiedler_index])
print("Fiedler index:", fiedler_index)

# =========================
# Swarm plot (horizontal jitter)
# =========================
rng = np.random.default_rng(42)
y_jitter = rng.uniform(-0.25, 0.25, size=n)

plt.figure(figsize=(9, 4))
plt.scatter(fiedler, y_jitter, s=18, alpha=0.7)
plt.axvline(0, linewidth=1)
plt.yticks([])
plt.title("Fiedler Vector Components (Swarm Plot)")
plt.xlabel("Fiedler component value")
plt.show()

# =========================
# 2-way clustering by sign of Fiedler vector
# =========================
cluster = np.where(fiedler < 0, 0, 1)

neg_count = int(np.sum(cluster == 0))
pos_count = int(np.sum(cluster == 1))
print("Negative cluster size:", neg_count)
print("Positive cluster size:", pos_count)

# =========================
# Graph visualization colored by cluster (reuse 'pos' from your part a)
# =========================
# If you don't already have pos, uncomment this:
# pos = nx.spring_layout(G, seed=42, k=1/np.sqrt(n))

plt.figure(figsize=(10, 10))

nx.draw_networkx_nodes(
    G,
    pos,
    node_size=20,
    node_color=cluster,   # 0/1 mapped to colormap
    cmap=plt.cm.Set2,
    alpha=0.9
)

nx.draw_networkx_edges(
    G,
    pos,
    width=0.3,
    alpha=0.25
)

plt.title("Movie Network Colored by Fiedler Sign (2 Clusters)")
plt.axis("off")
plt.show()
```
