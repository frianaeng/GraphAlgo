## Data wrangling 

```python
from collections import defaultdict
import pickle
import os

class MovieNetwork:
    def __init__(self):
        """Creates a movie similarity network"""
        # adjacency: movie -> {other_movie : weight}
        self.adjacency = defaultdict(dict)
        self.features = {}

    def getNumMovies(self):
        return len(self.adjacency)

    def getNumEdges(self):
        numEdges = 0
        for _, nbrs in self.adjacency.items():
            numEdges += len(nbrs)
        return int(numEdges / 2)

    def addMovie(self, movie, feature_set):
        """Add a movie and its feature set"""
        if movie not in self.adjacency:
            self.adjacency[movie] = {}
        self.features[movie] = set(feature_set)

    def addEdge(self, movie1, movie2, weight):
        """Add an undirected weighted edge"""
        self.adjacency[movie1][movie2] = weight
        self.adjacency[movie2][movie1] = weight


    def jaccard(self, s1, s2):
        if len(s1 | s2) == 0:
            return 0.0
        return len(s1 & s2) / len(s1 | s2)

    def buildEdges(self, threshold=0.0):
        """Compute Jaccard weights for all movie pairs"""
        movies = list(self.features.keys())

        for i in range(len(movies)):
            for j in range(i + 1, len(movies)):
                m1, m2 = movies[i], movies[j]
                w = self.jaccard(self.features[m1], self.features[m2])
                if w > threshold:
                    self.addEdge(m1, m2, w)

   

    def BFS(self, start):
        import queue
        Q = queue.SimpleQueue()
        Q.put(start)

        visited = {m: False for m in self.adjacency}
        parent  = {m: None  for m in self.adjacency}

        visited[start] = True

        while not Q.empty():
            u = Q.get()
            for v in self.adjacency[u]:
                if not visited[v]:
                    visited[v] = True
                    parent[v] = u
                    Q.put(v)
        return parent

    def countConnectedComponents(self):
        visited = {m: False for m in self.adjacency}
        components = 0

        for movie in self.adjacency:
            if not visited[movie]:
                components += 1
                stack = [movie]
                visited[movie] = True

                while stack:
                    u = stack.pop()
                    for v in self.adjacency[u]:
                        if not visited[v]:
                            visited[v] = True
                            stack.append(v)
        return components

    # -----------------------------
    # Similarity-specific queries
    # -----------------------------

    def mostSimilar(self, movie, k=5):
        """Return top-k most similar movies"""
        return sorted(
            self.adjacency[movie].items(),
            key=lambda x: x[1],
            reverse=True
        )[:k]

    # -----------------------------
    # Persistence
    # -----------------------------

    def save(self, path):
        os.makedirs(os.path.dirname(path), exist_ok=True)
        with open(path, "wb") as f:
            pickle.dump(self, f)


            

#A very high Jaccard weight means the two movies share most of their features and are very similar under 
#the chosen feature set.


    def loadDataset(self, filename):
        """
        Loads movie data and builds a movie similarity network
        based on Jaccard similarity of feature sets.
        """
        import pickle
    
        with open(filename, 'rb') as f:
            data = pickle.load(f)
    
        # data expected format:
        # movie -> iterable of features (e.g., actors, genres, etc.)
    
        # Step 1: add movies and feature sets
        for movie, features in data.items():
            self.addMovie(movie, features)
    
        movies = list(self.features.keys())
    
        # Step 2: compute pairwise Jaccard similarity
        for i in range(len(movies)):
            for j in range(i + 1, len(movies)):
                m1, m2 = movies[i], movies[j]
                s1, s2 = self.features[m1], self.features[m2]
    
                sim = self.jaccard(s1, s2)
                if sim > 0:
                    self.addEdge(m1, m2, sim)


    def visualize_network(self):
        """Visualize the movie similarity network using NetworkX and Matplotlib"""
        import networkx as nx
        import matplotlib.pyplot as plt
    
        G = nx.Graph()
    
        # Add nodes (movies)
        for movie in self.adjacency:
            G.add_node(movie)
    
        # Add edges (movie similarity, weighted)
        for movie, neighbors in self.adjacency.items():
            for other_movie, weight in neighbors.items():
                if not G.has_edge(movie, other_movie):
                    G.add_edge(movie, other_movie, weight=weight)
    
        # Layout
        pos = nx.circular_layout(G)
    
        # Draw graph
        plt.figure(figsize=(10, 8))
        nx.draw(
            G,
            pos,
            with_labels=True,
            node_color="lightgreen",
            node_size=500,
            font_size=8,
            font_weight="bold",
            edge_color="gray"
        )
    
        plt.title("Movie Similarity Network Visualization")
        plt.show()
    
```

```python
network = MovieNetwork()
    
# Example data: adding movies manually (replace with loadDataSet for file input)
network.addMovie("Movie1", {"Actor1", "Actor2"})
network.addMovie("Movie2", {"Actor2", "Actor4"})
network.addMovie("Movie3", {"Actor3", "Actor5"})
```

```python
movies = list(network.features.keys())

for i in range(len(movies)):
    for j in range(i + 1, len(movies)):
        m1, m2 = movies[i], movies[j]
        sim = network.jaccard(
            network.features[m1],
            network.features[m2]
        )
        if sim > 0:
            network.addEdge(m1, m2, sim)

network.visualize_network()
```

```python
# Example: explicit Jaccard similarity between two movies

movie1 = "Movie1"
movie2 = "Movie2"

features1 = network.features[movie1]
features2 = network.features[movie2]

intersection = features1 & features2
union = features1 | features2

jaccard_similarity = len(intersection) / len(union)

print("Movie 1 actors:", features1)
print("Movie 2 actors:", features2)
print("Jaccard similarity:", jaccard_similarity)
```

```python
import pickle
import types
import sys
from collections import defaultdict

def load_criterion_movie_features(path, num_movies=1805):
    """
    Returns:
        movie_features: dict[movie_id] -> set(actors)
    """

    # Stub ActorNetwork so pickle loads cleanly
    module = types.ModuleType("ActorNetwork")
    class ActorNetwork:
        pass
    module.ActorNetwork = ActorNetwork
    sys.modules["ActorNetwork"] = module

    with open(path, "rb") as f:
        actor_net = pickle.load(f)

    movie_features = defaultdict(set)

    # actor_net.movies: actor -> set(movie_ids)
    for actor, movie_ids in actor_net.movies.items():
        for m in movie_ids:
            movie_features[m].add(actor)

    # Ensure full 0..1804 movie index space exists
    for m in range(num_movies):
        movie_features[m]  # touch key

    return movie_features
```

```python
#from movie_network import MovieNetwork  # or wherever your class lives

CRITERION_PATH = "data/pkl/criterion_actor_network.pkl"

# Step 1: load movie -> actor features
movie_features = load_criterion_movie_features(CRITERION_PATH)

# Step 2: create MovieNetwork
MN = MovieNetwork()

# Step 3: add movies + feature sets
for movie_id, actors in movie_features.items():
    MN.addMovie(movie_id, actors)

print("Movies added:", MN.getNumMovies())

# Step 4: build edges using Jaccard similarity
# threshold=0 keeps all shared-actor connections
MN.buildEdges(threshold=0.0)

print("Edges added:", MN.getNumEdges())
print("Connected components:", MN.countConnectedComponents())
```
