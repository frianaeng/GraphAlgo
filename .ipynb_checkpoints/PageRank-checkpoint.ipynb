```python
import numpy as np
import time
import math
from ActorNetwork import ActorNetwork
from ActorMovieNetwork import ActorMovieNetwork
from MovieNetwork import MovieNetwork
import pickle
#!conda install -c conda-forge python-igraph -y

class PageRank:
    def page_rank(self,adj_matrix, damping=0.85, tol=1e-8, max_iter=100,record_entropy=True):
        start=time.time()
        n = adj_matrix.shape[0]
    
        A = adj_matrix.astype(float)
    
        # Handle dangling nodes
        row_sums = A.sum(axis=1)
        for i in range(n):
            if row_sums[i] == 0:
                A[i] = np.ones(n) / n
    
        # Row-normalize
        A = A / A.sum(axis=1, keepdims=True)
        M = A.T
    
        pr = np.ones(n) / n
        teleport = np.ones(n) / n
    
        entropies = []
      #  print("P0")
       # print(pr)
        for _ in range(max_iter):
            # pr_i -> pr_{i+1}
            new_pr = damping * (M @ pr) + (1 - damping) * teleport
            new_pr = new_pr / new_pr.sum()
         #   print(new_pr)
            # THIS is the answer to your question
            #entropies.append(cross_entropy(pr, new_pr))
            if record_entropy:
                entropies.append(cross_entropy(pr, new_pr))
           
            # Early stopping disabled to observe convergence over fixed iterations
            pr = new_pr

            # if np.linalg.norm(new_pr - pr, 1) < tol:
            #     pr = new_pr
            #     break
    
            
        end=time.time()
        print(end-start)
        return pr, entropies

```

```python
def build_adjacency_matrix_from_network(an):
    actors = list(an.costars.keys())
    n = len(actors)

    actor_to_index = {actors[i]: i for i in range(n)}
    A = np.zeros((n, n), dtype=int)

    for actor, neighbors in an.costars.items():
        i = actor_to_index[actor]
        for neighbor in neighbors:
            j = actor_to_index[neighbor]
            A[i][j] = 1
            A[j][i] = 1

    return A, actor_to_index

```

```python
def actor_network_to_matrix(actor_net):
    actors = list(actor_net.costars.keys())
    idx = {a:i for i,a in enumerate(actors)}
    n = len(actors)
    A = np.zeros((n,n))

    for a, neighbors in actor_net.costars.items():
        i = idx[a]
        for b in neighbors:
            j = idx[b]
            A[i,j] = 1

    return A, actors
```

```python
def actor_movie_to_matrix(am_net):
    actors = list(am_net.actor_to_movies.keys())
    movies = list(am_net.movie_to_actors.keys())

    nodes = actors + movies
    idx = {n:i for i,n in enumerate(nodes)}

    n = len(nodes)
    A = np.zeros((n,n))

    for actor, movs in am_net.actor_to_movies.items():
        for movie in movs:
            i = idx[actor]
            j = idx[movie]
            A[i,j] = 1
            A[j,i] = 1   # undirected

    return A, nodes
```

```python
import numpy as np

def movie_network_to_matrix(mn: MovieNetwork):
    """
    Convert MovieNetwork adjacency dict into a weighted adjacency matrix.

    Rows = source movie
    Columns = destination movie
    Values = Jaccard similarity weights
    """
    movies = list(mn.adjacency.keys())
    index = {movie: i for i, movie in enumerate(movies)}
    n = len(movies)

    A = np.zeros((n, n))

    for m1, nbrs in mn.adjacency.items():
        i = index[m1]
        for m2, weight in nbrs.items():
            j = index[m2]
            A[i, j] = weight

    return A, movies
```

```python
import math
import numpy as np

# Define the four probability vectors
P = [
    np.array([1.0, 0.0, 0.0]),
    np.array([0.0, 1.0, 0.0]),
    np.array([1/3, 1/3, 1/3]),
    np.array([1/4, 1/4, 1/2]),
]

labels = ["p1=[1,0,0]", "p2=[0,1,0]", "p3=[1/3,1/3,1/3]", "p4=[1/4,1/4,1/2]"]
def cross_entropy(p, q, base=2):
    """
    Cross-entropy H(p, q) = - sum_i p_i log q_i
    Returns inf if q assigns zero probability where p > 0.
    """
    H = 0.0
    for pi, qi in zip(p, q):
        if pi == 0:
            continue
        if qi == 0:
            return math.inf
        H -= pi * (math.log(qi) / math.log(base))
    return H

# Build the 4x4 table
table = np.zeros((4, 4), dtype=float)

for i in range(4):
    for j in range(4):
        table[i, j] = cross_entropy(P[i], P[j])

# Pretty print
print("Cross-Entropy Table (bits)\n")
print(" " * 18 + "  ".join(f"{lbl:>18}" for lbl in labels))
for i, row in enumerate(table):
    row_str = []
    for val in row:
        if math.isinf(val):
            row_str.append(f"{'inf':>18}")
        else:
            row_str.append(f"{val:18.3f}")
    print(f"{labels[i]:>18}  " + "  ".join(row_str))
```

```python
#Run and time PageRank using NetworkX.
import networkx as nx
import numpy as np
import time

def pagerank_networkx(adj_matrix, damping=0.85):
    """
    Runs and times PageRank using NetworkX.

    Parameters:
        adj_matrix (np.ndarray): adjacency matrix
        damping (float): damping factor (alpha)

    Returns:
        pr_dict (dict): node -> PageRank score
        elapsed (float): runtime in seconds
    """
    n = adj_matrix.shape[0]

    # Create directed graph from adjacency matrix
    G = nx.DiGraph()
    for i in range(n):
        for j in range(n):
            if adj_matrix[i, j] != 0:
                G.add_edge(i, j, weight=adj_matrix[i, j])
    print("Nodes:", G.number_of_nodes())
    print("Edges:", G.number_of_edges())

    start = time.time()
    pr_dict = nx.pagerank(G, alpha=damping, weight="weight")
    elapsed = time.time() - start

    return pr_dict, elapsed
```

```python
#Run and time PageRank using iGraph.

import igraph as ig
import numpy as np
import time

def igraph_pagerank(adj_matrix, damping=0.85):
    """
    Runs PageRank using iGraph and returns (pagerank_vector, runtime_seconds)
    """
    n = adj_matrix.shape[0]

    edges = []
    weights = []

    for i in range(n):
        for j in range(n):
            if adj_matrix[i, j] > 0:
                edges.append((i, j))
                weights.append(adj_matrix[i, j])

    g = ig.Graph(n=n, edges=edges, directed=True)
    g.es["weight"] = weights

    start = time.time()
    pr = g.pagerank(damping=damping, weights="weight")
    end = time.time()

    return np.array(pr), end - start
```
