```python
import numpy as np

def adjacency_to_incidence(A):
    n = A.shape[0]
    edges = []

    # extract edges from upper triangle
    for i in range(n):
        for j in range(i + 1, n):
            if A[i, j] != 0:
                edges.append((i, j))

    m = len(edges)
    B = np.zeros((n, m), dtype=int)

    for k, (i, j) in enumerate(edges):
        B[i, k] = 1
        B[j, k] = -1

    return B, edges
```

```python
# avoid division by zero
degrees = arr.sum(axis=1)
n = arr.shape[0]
deg_inv_sqrt = np.zeros_like(degrees)
deg_inv_sqrt[degrees > 0] = 1.0 / np.sqrt(degrees[degrees > 0])

D_inv_sqrt = np.diag(deg_inv_sqrt)

L_sym = np.eye(n) - D_inv_sqrt @ arr @ D_inv_sqrt
```

```python
import numpy as np

# eigen-decomposition
eigvals, eigvecs = np.linalg.eigh(L_sym)

# first two eigenvalues
lambda_1 = eigvals[0]
lambda_2 = eigvals[1]

# corresponding eigenvectors
v1 = eigvecs[:, 0]
v2 = eigvecs[:, 1]

lambda_1, lambda_2
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Fiedler vector (second smallest eigenvalue)
fiedler = v2

n = len(fiedler)

# vertical jitter to prevent overlap
rng = np.random.default_rng(0)
y_jitter = rng.uniform(-0.15, 0.15, size=n)

plt.figure(figsize=(10, 3))

# swarm points (horizontal)
plt.scatter(
    fiedler,
    y_jitter,
    s=20,
    alpha=0.7
)

# vertical line at x = 0 (natural cut)
plt.axvline(0, linestyle="--", linewidth=2)

plt.yticks([])  # no y-axis meaning
plt.xlabel("Fiedler Vector Component Value")
plt.title("Horizontal Swarm Plot of Fiedler Vector")

plt.tight_layout()
plt.show()
```

```python
import numpy as np

def louvain(A, seed=0, max_iter=100):
    """
    Louvain community detection (from scratch).

    Input:
        A : (n x n) adjacency matrix (undirected, weighted or unweighted)

    Output:
        communities : array of community labels for each node
    """

    rng = np.random.default_rng(seed)
    A = A.astype(float)
    n = A.shape[0]

    # =========================
    # Phase 1: Local Moving
    # =========================

    # Step 1: each node starts in its own community
    communities = np.arange(n)

    degrees = A.sum(axis=1)
    m2 = A.sum()  # equals 2m for undirected graph

    def modularity_gain(i, target_comm):
        k_i = degrees[i]

        k_i_in = 0.0
        tot_c = 0.0

        for j in range(n):
            if communities[j] == target_comm:
                k_i_in += A[i, j]
                tot_c += degrees[j]

        return (k_i_in / m2) - (k_i * tot_c) / (m2 * m2)

    moved = True
    iteration = 0

    while moved and iteration < max_iter:
        moved = False
        iteration += 1

        nodes = np.arange(n)
        rng.shuffle(nodes)

        for i in nodes:
            current_comm = communities[i]

            neighbor_comms = set()
            for j in range(n):
                if A[i, j] > 0:
                    neighbor_comms.add(communities[j])

            best_comm = current_comm
            best_gain = 0.0

            for c in neighbor_comms:
                gain = modularity_gain(i, c)
                if gain > best_gain:
                    best_gain = gain
                    best_comm = c

            if best_comm != current_comm:
                communities[i] = best_comm
                moved = True

    # =========================
    # Phase 2: Aggregation
    # =========================

    # Relabel communities to consecutive IDs
    unique_comms = np.unique(communities)
    comm_map = {c: idx for idx, c in enumerate(unique_comms)}
    communities = np.array([comm_map[c] for c in communities])

    num_comms = len(unique_comms)

    # If no aggregation occurred, stop
    if num_comms == n:
        return communities

    # Build aggregated adjacency matrix
    A_new = np.zeros((num_comms, num_comms))

    for i in range(n):
        for j in range(n):
            ci = communities[i]
            cj = communities[j]
            A_new[ci, cj] += A[i, j]

    # Recurse on aggregated graph
    higher_level = louvain(A_new, seed=seed)

    # Lift labels back to original nodes
    final_labels = np.zeros(n, dtype=int)
    for i in range(n):
        final_labels[i] = higher_level[communities[i]]

    return final_labels
```
